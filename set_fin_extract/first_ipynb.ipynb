```python
from io import StringIO
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
url = "https://www.set.or.th/en/market/product/stock/quote/ADVANC/financial-statement/company-highlights"
driver.get(url)
```

```python
html = driver.page_source
data_list = pd.read_html(StringIO(html))
df1 = data_list[0]
df2 = data_list[1]
```

```python
df1 = df1[~df1["Period as of"].isin(["Financial Data", "Financial Ratio"])]
header_row = pd.DataFrame([df1.columns.tolist()], columns=df1.columns)
df1_demoted = pd.concat([header_row, df1], ignore_index=True)
df1_demoted.columns = [1,2,3,4,5,6]
df1_demoted
```

```python
header_row = pd.DataFrame([df2.columns.tolist()], columns=df2.columns)
df2_demoted = pd.concat([header_row, df2], ignore_index=True)
df2_demoted.columns = [1,2,3,4,5,6]
df2_demoted
```

```python
df_combined = pd.concat([df1_demoted, df2_demoted], ignore_index=True)
df_trans = df_combined.T
df_trans
```

```python
df_trans.columns = df_trans.iloc[0]
df_trans = df_trans.drop(index=1).reset_index(drop=True)
```

```python
df_trans["Period as of"] = df_trans["Period as of"].apply(lambda x: " ".join(str(x).split()[-3:]))
date_cols = ["Statistics as of", "F/S Period (As of date)", "Period as of"]
for col in date_cols:
    df_trans[col] = pd.to_datetime(df_trans[col], format="%d %b %Y")
df_trans
```

```python
df_trans["Stock"] = "ADVANC"
print(df_trans.iloc[0])
```
